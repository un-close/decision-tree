```python
import numpy as np
flower = [3.2, 2.2, 2.5, 1.9]
sq_sum = 0
for i in range(4):
    sq_sum += (X.iloc[:, i] - flower[i]) ** 2 
Sum = np.sqrt(sq_sum)
Sum
```

```python
X['dist'] = Sum
X['Species'] = y
X.sort_values('dist')
```

```python
X.sort_values('dist').head(5).Species.value_counts()
```

```python
iris['target_names'][X.sort_values('dist').head(5).Species.value_counts()[0]]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
x = pd.DataFrame(iris['data'],columns= iris['feature_names'])
y = iris['target']
x_train, x_test, y_train, y_test = train_test_split(x, y)
model_1 = KNeighborsClassifier()
model_1.fit(x_train, y_train)

```

```python
print(model_1.score(x_train, y_train))
print(model_1.score(x_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, model_1.predict(x_test)))
```

```python
max_score = float('-inf')
optimal = 5
scores = list()
for k in range(5, 12):
    model_1 = KNeighborsClassifier(n_neighbors=k)
    model_1.fit(x_train, y_train)
    score = model_1.score(x_test, y_test)
    scores.append(score)
    if max_score < score:
        max_score, optimal = score, k
print(optimal)

```

```python
import matplotlib.pyplot as plt
plt.plot(range(5,12), scores)
plt.show()
```
